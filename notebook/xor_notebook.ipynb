```python
import numpy as np
from neurokit import *
```

```python
# XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

```python
# Model
class XORNet(Module):
    def __init__(self):
        super().__init__()
        self.fc1 = Linear(2, 4)
        self.act = Sigmoid()
        self.fc2 = Linear(4, 1)
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.act(x)
        x = self.fc2(x)
        x = self.act(x)
        return x
```

```python
model = XORNet()
criterion = MSELoss()
optimizer = Adam(model.parameters(), lr=0.1)
```

```python
# Training
for epoch in range(1000):
    # Forward
    pred = model(Tensor(X))
    loss = criterion(pred, Tensor(y))
    
    # Backward
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    
    if epoch % 200 == 0:
        print(f"Epoch {epoch}, Loss: {loss.data:.4f}")
```

```python
# Test
print("\nTest Results:")
pred = model(Tensor(X))
for i in range(4):
    print(f"{X[i]} -> {pred.data[i, 0]:.4f} (target: {y[i, 0]})")
```
